<h2>Get the numerical variables from our dataset.<h2>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_csv('./files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
def prepare_df(df):
    df = df.drop_duplicates()
    df = df.reset_index(drop = True)
    df.columns = [col.lower().replace(' ', '_') for col in df.columns]
    df = round(df,2)
    return df
```

```python
data = prepare_df(data)
```

```python
numericals = data.select_dtypes(np.number)
```

```python
numericals.info()
```

<br><h2>Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.<h2>

```python
fig, axes = plt.subplots(4,2,figsize=(20,20), dpi=300)
sns.distplot(x=numericals['customer_lifetime_value'], ax=axes[0,0])
axes[0,0].set_xlabel('Customer Lifetime Value')
sns.distplot(x=numericals['income'], ax=axes[0,1])
axes[0,1].set_xlabel('Income')
sns.distplot(x=numericals['monthly_premium_auto'], ax=axes[1,0])
axes[1,0].set_xlabel('Monthly Premium Auto')
sns.distplot(x=numericals['months_since_last_claim'], ax=axes[1,1])
axes[1,1].set_xlabel('Months Since Last Claim')
sns.distplot(x=numericals['months_since_policy_inception'], ax=axes[2,0])
axes[2,0].set_xlabel('Months Since Policy Inception')
sns.distplot(x=numericals['number_of_open_complaints'], ax=axes[2,1])
axes[2,1].set_xlabel('Number of Open Complaints')
sns.distplot(x=numericals['number_of_policies'], ax=axes[3,0])
axes[3,0].set_xlabel('Number of Policies')
sns.distplot(x=numericals['total_claim_amount'], ax=axes[3,1])
axes[3,1].set_xlabel('Total Claim Amount')
```

One graphs looks a bit like a normal distribution (months since policy inception) but very wide, some resemble an exponential distribution:<br>
-Customer lifetime value (the most).<br>
-Number of open complaints.<br>
-Monthly premium auto.<br>
-Total claim amount (the lesser).<br>

<br><h2>Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.<h2>

I will check the log transformation on the 4 columns mentioned before. No negative values on them so I can apply it directly.

```python
def log_transfom(x):
    x = np.log10(x)
    if np.isfinite(x):
        return x
    else:
        return np.NAN
    
def log_transfom2(x):
    x = np.log10(x) 
    if np.isfinite(x):
        return x
    else:
        return 0
```

```python
num_median = numericals.copy()
num_mean = numericals.copy()
```

```python
num_median['customer_lifetime_value'] = list(map(log_transfom, num_median['customer_lifetime_value']))
num_median['customer_lifetime_value'] = num_median['customer_lifetime_value'].fillna(np.median(num_median['customer_lifetime_value']))
num_mean['customer_lifetime_value'] = list(map(log_transfom, num_mean['customer_lifetime_value']))
num_mean['customer_lifetime_value'] = num_mean['customer_lifetime_value'].fillna(np.median(num_mean['customer_lifetime_value']))

num_median['number_of_open_complaints'] = list(map(log_transfom2, num_median['number_of_open_complaints']))
num_median['number_of_open_complaints'] = num_median['number_of_open_complaints'].fillna(np.median(num_median['number_of_open_complaints']))
num_mean['number_of_open_complaints'] = list(map(log_transfom2, num_mean['number_of_open_complaints']))
num_mean['number_of_open_complaints'] = num_mean['number_of_open_complaints'].fillna(np.median(num_mean['number_of_open_complaints']))

num_median['monthly_premium_auto'] = list(map(log_transfom, num_median['monthly_premium_auto']))
num_median['monthly_premium_auto'] = num_median['monthly_premium_auto'].fillna(np.median(num_median['monthly_premium_auto']))
num_mean['monthly_premium_auto'] = list(map(log_transfom, num_mean['monthly_premium_auto']))
num_mean['monthly_premium_auto'] = num_mean['monthly_premium_auto'].fillna(np.median(num_mean['monthly_premium_auto']))

num_median['total_claim_amount'] = list(map(log_transfom, num_median['total_claim_amount']))
num_median['total_claim_amount'] = num_median['total_claim_amount'].fillna(np.median(num_median['total_claim_amount']))
num_mean['total_claim_amount'] = list(map(log_transfom, num_mean['total_claim_amount']))
num_mean['total_claim_amount'] = num_mean['total_claim_amount'].fillna(np.median(num_mean['total_claim_amount']))
```

```python
fig, axes = plt.subplots(4,2,figsize=(20,20), dpi=300)
sns.distplot(x=num_median['customer_lifetime_value'], ax=axes[0,0])
axes[0,0].set_xlabel('Customer Lifetime Value - median')
sns.distplot(x=num_mean['customer_lifetime_value'], ax=axes[0,1])
axes[0,1].set_xlabel('Customer Lifetime Value - mean')
sns.distplot(x=num_median['monthly_premium_auto'], ax=axes[1,0])
axes[1,0].set_xlabel('Monthly Premium Auto - median')
sns.distplot(x=num_mean['monthly_premium_auto'], ax=axes[1,1])
axes[1,1].set_xlabel('Monthly Premium Auto - mean')
sns.distplot(x=num_median['number_of_open_complaints'], ax=axes[2,0])
axes[2,0].set_xlabel('Number of Open Complaints - median')
sns.distplot(x=num_mean['number_of_open_complaints'], ax=axes[2,1])
axes[2,1].set_xlabel('Number of Open Complaints - mean')
sns.distplot(x=num_median['total_claim_amount'], ax=axes[3,0])
axes[3,0].set_xlabel('Total Claim Amount - median')
sns.distplot(x=num_mean['total_claim_amount'], ax=axes[3,1])
axes[3,1].set_xlabel('Total Claim Amount - mean')
```

Customer lifetime value is more centered but still doesn't resemble a normal distribution. Also, the other columns have shifted towards the right so even if they looked a bit like an exponential distribution, after the transformation is clear that they weren't really.
